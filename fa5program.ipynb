### Objective:
Create a function to calculate the total price of items. This program will ask the user for the price of one item and the quantity, then display the calculated total.

```python
def calculate_total_price(price_per_item, quantity):
    """
    Calculates the total price of items.

    Args:
        price_per_item (float): The price of a single item.
        quantity (int): The number of items purchased.

    Returns:
        float: The total calculated price.
    """
    return price_per_item * quantity

# Get input from the user
try:
    price = float(input("Enter the price of one item: "))
    num_items = int(input("Enter the number of items purchased: "))

    # Validate inputs
    if price < 0 or num_items < 0:
        print("Price and quantity cannot be negative. Please enter positive values.")
    else:
        # Calculate the total price using the function
        total_cost = calculate_total_price(price, num_items)

        # Display the result
        print(f"The total price for {num_items} items at ${price:.2f} each is: ${total_cost:.2f}")
except ValueError:
    print("Invalid input. Please enter valid numbers for price and quantity.")
```
